# HYPOTHESIS TESTING

### 1. State the Hypotheses statement

#### Data Provided:
#### ●	The theoretical weekly operating cost model: W = $1,000 + $5X
#### ●	Sample of 25 restaurants with a mean weekly cost of Rs. 3,050
#### ●	Number of units produced in a week (X) follows a normal distribution with a mean (μ) of 600 units and a standard deviation (σ) of 25 units


#### W = $1000 + $5X
#### for a week the W will be 
####     W = $1000+$5X
####       = $1000+$5*600
####       = 4000

### so:
#### H0 = The weekly operating costs are higher than 4000
#### Ha = The weekly opearting costs are less than or equal to 4000 


### 2.Calculate the Test Statistic:
#### Use the following formula to calculate the test statistic(z):
#### where:
#### ●	ˉxˉ = sample mean weekly cost (Rs. 3,050)
#### ●	μ = theoretical mean weekly cost according to the cost model (W = $1,000 + $5X for X = 600 units)
#### ●	σ = 5*25 units
#### ●	n = sample size (25 restaurants)


```python
import numpy as np
import scipy.stats as stats
# z-tests
import statsmodels.api as sm
```

```python
sample_mean = 3050
alpha = 0.05

# known population values
pop_std = 125
n = 25
pop_mean = 4000

# one tailed Z-test
z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p_value = 1-(stats.norm.cdf(z_score))

print(f"Z-score: {z_score:.3f}")
print(f"P-value: {p_value:.3f}")


```

```python

from scipy.stats import chi2

alpha = 0.05

# Observed frequencies (assumed categorical grouping)
observed = np.array([8, 10, 7])

# Expected frequencies (uniform distribution assumption)
expected = np.array([25/3, 25/3, 25/3])

# Chi-square statistic
chi_square_stat = ((observed - expected) ** 2 / expected).sum()

# Degrees of freedom
df = len(observed) - 1

# Critical value
chi_critical = chi2.ppf(1 - alpha, df)

# P-value
p_value = 1 - chi2.cdf(chi_square_stat, df)

print("Chi-Square Statistic:", chi_square_stat)
print("Critical Value:", chi_critical)
print("P-value:", p_value)

if chi_square_stat > chi_critical:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

### 3.Determine the Probability and compare:
#### Using the alpha level of 5% (α = 0.05),
### 4. Make a Decision:
#### Compare the test statistic with the critical value to decide whether to reject the null hypothesis


```python
if p_value < alpha:
    print("Reject the null hypothesis: the weekly operating costs are higher than 4000.")
else:
    print("Fail to reject the null hypothesis: There is not enough evidence to conclude that the weekly operating costs are higher than 4000") 
```

### 5. Conclusion:
#### Based on the decision in step 4, conclude whether there is strong evidence to support the restaurant owners' claim that the weekly operating costs are higher than the model suggests.

#### The conclusion is:
#### There is no strong evidence to support the restaurant owners’ claim that weekly operating costs are higher than the model suggests.
#### From step 4, the test statistic z is far below the critical value for a right‑tailed test at alpha = 0.05, so the null hypothesis (mean cost equals the model value) is not rejected. This means the sample data do **not** show a statistically significant increase in weekly operating costs relative to the theoretical model, and therefore the claim that costs are higher than the model suggests is not supported at the 5% significance level.
